# **Business Questions**

## **1) What is the mos popular realse year for movies on netflix**

```python
#Import Libraries
import pandas as pd
```

```python
#Data Frame
df = pd.read_csv("/content/netflix_titles.csv")
df.head(5)
```

```python
movie_df = df[df["type"] =="Movie"]
```

```python
#Check data
df["type"].unique()
```

```python
movie_df["count"]= 1
summed_release_year = movie_df.groupby(["release_year"]).sum().reset_index()[["release_year","count"]]
```

```python
summed_release_year.sort_values(["count"],ascending=False)
```

## **2)What year did netflix add  the most content to its plataform**

```python
df.head(5)
```

```python
# Add New Column
df["new_date"] = pd.to_datetime(df["date_added"])
df["new_date"].info()
df["new_date"].head()
```

```python
df["date_added_year"] = df["new_date"].dt.year
df.head(5)
```

```python
df["count"]= 1
summed_date_added_year = df.groupby(["date_added_year"]).sum().reset_index()[["date_added_year","count"]]
```

```python
summed_date_added_year.sort_values(["count"],ascending=False)
```

## **3)What is the most popular month to add new content**

```python
df.head()
```

```python
df["date_added_month"]= df["new_date"].dt.month
df.head()
```

```python
df["count"]= 1
date_added_month = df.groupby(["date_added_month"]).sum().reset_index()[["date_added_month","count"]]
```

```python
date_added_month.sort_values(["count"],ascending=False)
```

## **5) Which actor shows up most frequently**

```python
df.head(5)
```

```python
df[["title","cast"]]
```

```python
title_cast_df = df[["title","cast"]]
title_cast_df = title_cast_df.dropna()
```

```python
cast_names_counter = {}

for index, row in title_cast_df.iterrows():
    movie_cast = row["cast"]
    movie_cast_split = movie_cast.split(".")
    movie_cast_stripped = {name.strip().lower() for name in movie_cast_split}
    for name in movie_cast_stripped:
        cast_names_counter[name] = cast_names_counter.get(name, 0) + 1

sorted_cast_names = dict(sorted(cast_names_counter.items(), key=lambda item: item[1], reverse=True))
print(sorted_cast_names)


```
